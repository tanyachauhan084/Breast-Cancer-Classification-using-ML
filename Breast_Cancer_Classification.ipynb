```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
patient_data= pd.read_csv("data.csv")
patient_data
```

```python
patient_data.isnull().sum()
```

```python
patient_data= patient_data.drop('Unnamed: 32', axis=1)
```

```python
patient_data
```

```python
patient_data.isnull().sum()
```

```python
encoding = LabelEncoder()
encoding
```

```python
patient_data['diagnosis']= encoding.fit_transform(patient_data['diagnosis'])
patient_data['diagnosis']
```

```python
patient_data
```

```python
patient_data['diagnosis'].value_counts()
```

```python
X= patient_data.drop(['id','diagnosis'], axis=1)
X
```

```python
Y = patient_data['diagnosis']
Y
```

```python
corr = X.corrwith(Y)
corr
```

```python
standardscale = StandardScaler()
standardscale
```

```python
X =standardscale.fit_transform(X)
X
```

```python
model = LogisticRegression()
model
```

```python
X_train, X_test, Y_train,Y_test = train_test_split(X,Y, test_size= 0.1, stratify = Y, random_state= 42)
```

```python
model.fit(X_train,Y_train)
```

```python
prediction_X_train_data = model.predict(X_train)
accuracy_X_train_data = accuracy_score(prediction_X_train_data, Y_train)
accuracy_X_train_data*100
```

```python
prediction_X_test_data = model.predict(X_test)
accuracy_X_test_data= accuracy_score(prediction_X_test_data, Y_test)
accuracy_X_test_data*100
```
